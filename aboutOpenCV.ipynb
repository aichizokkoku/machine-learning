OpenCVについて

```
import cv2
```

download_imread.py

```
#画像のダウンロード
import urllib.request as req
url = "http://uta.pw/shodou/img/28/214.png"
req.urlretrieve(url, "test.png")

#OpenCVで読み込む
img = cv2.imread("test.png")
print(img)
```

画像をインライン表示できるようにする

```
%matplotlib inline
```

imshow.py

```
#ダウンロードした画像を画面に表示する
import matplotlib.pyplot as plt
import cv2
img = cv2.imread("test.png")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```
img = cv2.imread("test.png")
plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

ここからはネガポジ反転、グレースケール変換などを試してみる

negaposi.py

```
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("test.png")

#ネガポジ反転
img = 255 - img

plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

gray.py

```
import matplotlib.pyplot as plt
import cv2
img = cv2.imread("test.png")

plt.axis("off")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY))
plt.show()
```
